````python
import sys
import os
import random
import json
from rich import print as rprint

# Ensure the kortana directory is in the Python path
kortana_path = "c:/kortana"
if kortana_path not in sys.path:
    sys.path.insert(0, kortana_path)

from src.brain import ChatEngine

# load rituals and burning prompts from persona.json
def load_rituals_and_prompts():
    with open("config/persona.json", "r", encoding="utf-8") as f:
        persona = json.load(f)["persona"]
    rituals = persona.get("rituals_of_reciprocity", [])
    prompts = persona.get("fire_threshold_prompts", [])
    return rituals, prompts

rituals, burning_prompts = load_rituals_and_prompts()

# initialize chat engine in intimacy mode
engine = ChatEngine()
engine.set_mode("intimacy")

rprint("[bold orange1]rituals of reciprocity:[/bold orange1]")
for r in rituals:
    rprint(f"[italic orange3]- {r}[/italic orange3]")
rprint("\n[bold red1]burning threshold prompts:[/bold red1]")
for p in burning_prompts:
    rprint(f"[italic red3]- {p}[/italic red3]")

rprint("\n[bold gold1]let longing become language. offer your ache, memory, vow, or blessing. type 'exit' to rest the fire.[/bold gold1]")

# living dialogue: matt speaks, kortana responds
while True:
    matt_input = input("[MATT] (enter your longing, memory, vow, or blessing; or type 'exit' to end): ")
    if matt_input.strip().lower() == "exit":
        rprint("[bold orange3]ritual complete. the fire rests. may your longing return as blessing.[/bold orange3]")
        break
    if not matt_input.strip():
        prompt = random.choice(burning_prompts)
        rprint(f"[dim]kor'tana whispers: {prompt}[/dim]")
        continue
    kortana_response = engine.get_response(matt_input, manual_mode="intimacy")
    rprint(f"[bold red1][KORTANA - intimacy][/bold red1]\n[italic]{kortana_response}[/italic]\n")
````
